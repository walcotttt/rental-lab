## Data Exploration

Before cleaning our dataset, let's explore the features that our dataset contains, as well as visuals.

```python
# import modules

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load in  dataset 
df = pd.read_excel("../../data/raw/realestate.xlsx")

# print first 5 rows of the dataset 
df.head()
```

```python
# find the datatypes of each column
df.info()
```

```python
# print out the summary statistics of all numeric columns
df.describe()
```

# Univariate Exploratory Analysis

Exploring the distribution of single variables in this dataset.

```python
# histogram of "Y house price of unit area"
sns.histplot(df["Y house price of unit area"])
plt.show()
```

```python
# box plot of "Y house price of unit area"
sns.boxplot(df["Y house price of unit area"])
plt.show()
```

```python
# histogram of "X2 house age"
sns.histplot(df["X2 house age"])
plt.show()
```

```python
# boxplot of "X2 house age"
sns.boxplot(df["X2 house age"])
plt.show()
```

```python
# histogram of "X4 number of convenience stores"
sns.histplot(df["X4 number of convenience stores"])
plt.show()
```

```python
# boxplot of "X4 number of convenience stores"
sns.boxplot(df["X4 number of convenience stores"])
plt.show()
```

# Bivariate Exploratory Analysis

Exploring the relationship of multiple variables in this dataset.

```python
# scatter plot of "X2 house age" & "Y house price of unit area"
sns.scatterplot(data=df, x="X2 house age", y="Y house price of unit area")
plt.show()
```

```python
# scatter plot of "X4 number of convenience stores" and "Y house price of unit area"
sns.scatterplot(data=df, x="X4 number of convenience stores", y="Y house price of unit area")
plt.show()
```

```python
# correlation heat map of selected columns in the dataset
selected_cols = ["X2 house age", "X4 number of convenience stores", "Y house price of unit area"]

mask = np.triu(np.ones_like(df[selected_cols].corr(), dtype=bool))

sns.heatmap(df[selected_cols].corr(), annot=True, mask=mask)
plt.show()
```
